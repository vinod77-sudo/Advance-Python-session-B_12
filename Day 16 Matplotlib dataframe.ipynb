```python
Compare the population of china and us:
========================================
```

```python
us=df[df.country=='United States']
```

```python
us
```

```python
china=df[df.country=='China']
```

```python
china
```

```python
plt.plot(us.year,us.population)
plt.xlabel('years')

plt.ylabel('population')

plt.title("population in United states between years 1952 -2007")

plt.legend(['The curve represents the population change'])  # le8 ,on left top means it can have 8 numbers

plt.show()
```

```python
plt.plot(china.year,china.population/10**6)  # here we are dividing the big number with 1 milion
plt.xlabel('years')

plt.ylabel('population')

plt.title("population in china between years 1952 -2007")

plt.legend(['The curve represents the population change'])  

plt.show()
```

```python
data=pd.read_csv('countries.csv')
```

```python
data[data.country=='South Africa']
```

```python

pop_spain=data[data.country=='Spain']

pop_mx=data[data.country=='Mexico']

pop_ind=data[data.country=='India']



pop_sou=data[data.country=='South Africa']

pop_net=data[data.country=='Netherlands']


plt.plot(pop_spain.year,pop_spain.population/10**6)

plt.plot(pop_mx.year,pop_mx.population/10**6)

plt.plot(pop_ind.year,pop_ind.population/10**6)

plt.plot(pop_sou.year,pop_sou.population/10**6)

plt.plot(pop_net.year,pop_net.population/10**6)


plt.xlabel('years')

plt.ylabel('population')

plt.title("population in between years 1952 -2007 of various countries")

plt.legend(['The curve represents the population change of spain',
            'The curve represents the population change of Mexico',
            'The curve represents the population change of India',
           'The curve represents the population change of south Africa',
           'The curve represents the population change of Netherlands'],loc="upper right")




plt.show()

```
