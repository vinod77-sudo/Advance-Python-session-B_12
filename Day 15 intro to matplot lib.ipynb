```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
x=[1,2,5]

y=[1,4,10]
```

```python
plt.plot(x,y)   # a normal plotting

plt.show()
```

```python
plt.plot(x,y)

plt.xlabel('Sales')

plt.ylabel('Revenues')

plt.show()
```

```python

plt.plot(x,y)        # adding lablels to the x and y to make it more clear an adding title to the description

plt.xlabel('Sales')

plt.ylabel('Revenues')

plt.title("The comparision of sales between years 19-20")

plt.show()

```

```python
z=[3,5,6]    # adding another array
```

```python
plt.plot(x,y)      # adding the legend for the clear idea of the curve represents

plt.xlabel("Sales")

plt.ylabel("Revenues")

plt.title("The comparision between the sales in years 19-20")

plt.legend(['This is comparision line'])


plt.show()
```

```python
plt.plot(x,y)      # Here plotted with the another array ,

plt.plot(x,z)

plt.xlabel('Sales')

plt.ylabel('Revenues')


plt.title('Comparision of sales and revenuse in the year 19-20')

plt.legend(['the comparision between x and y','the comparision between x and z'])

plt.show()
```

```python
sd=pd.read_csv('sample_data.csv')   # we are reading the the values from csv file

sd
```
